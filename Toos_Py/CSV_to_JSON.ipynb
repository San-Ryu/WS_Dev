```python
## CSV -> MongoDB
def csv_to_json(key):
    # CSV 파일 읽기
    try:
        df = pd.read_csv(dict_path[key][0], encoding='cp949')
    except UnicodeDecodeError:
        df = pd.read_csv(dict_path[key][0])
        
    # DataFrame을 JSON 형태로 변환
    records = df.to_dict(orient="records")
    
    # Atlas 연결 문자열 (예시)
    client = MongoClient(str_URI_MongoDB)
    db = client[dict_path[key][1]]          # 사용할 DB
    collection = db[str(key)]  # Collection 지정
    # 여러 문서 삽입
    collection.insert_many(records)
    print("데이터 삽입 완료!")
    client.close()
```
