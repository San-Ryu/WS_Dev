### BOJ1000 A+B
[링크] https://www.acmicpc.net/problem/1000

[문제]
두 정수 A와 B를 입력받은 다음, A+B를 출력하는 프로그램을 작성하시오.

[입력]
첫째 줄에 A와 B가 주어진다. (0 < A, B < 10)

[출력]
첫째 줄에 A+B를 출력한다.

[예제 입력 1]
1 2

[예제 출력 1]
3

[MEMO]
(1) 별도의 로직이 존재하지 않는다.
(초심자의 경우) Python 형식의 문제에서 어떻게 입력값을 받고, 결과값을 출력하는지에 더욱 중점을 두어 학습해야 한다.

(2) (번외) Java의 경우, int의 범위를 벗어난 숫자는 어떻게 해야할지 생각해봐야 한다. 하지만, Python은 Int가 고정된 크기를 갖지 않는다.
다만, 문제와 별도로 응용하자면 Float를 어떻게 처리할지는 생각해봐야 한다.
(문제에서는 "정수"를 명시했으므로 무관)

```python
## Int
a, b = map(int, input().split())
print(a + b)
```

```python
## Float
a, b = map(float, input().split())
print(a + b)
```
