[출력]  
첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한다.

3  
5  
1  
7

"BOJ2750_수정렬하기2"의 코드를 그대로 가져오면, 메모리 초과가 발생한다.  
입력의 크기가 최대 10,000,000개이고, 메모리 제한은 8MB이기에, 일반적인 Sort를 사용할 수 없다.  
따라서, Counting Sort를 사용하는 것이 방법이 될 수 있다.  

(1) 수의 범위에 따른 리스트[[0] * 10000] 생성  
(2) 입력값 배열을 한 번 읽고, Index에 해당하는 수를 읽을 때마다 해당 Index의 값을 +=1  
    예를 들어, 입력값 배열이 [3, 3, 2, 2, 5, 1]인 경우,  
    count[1] = 1,count[2] = 2, count[3] = 2, count[5] = 1  
    이 된다.  
(3) 입출력 최적화를 위해,    
    sys.stdin.readline()으로 입력   
    sys.stdout.write() 또는 '\n'.join()을 통해 일괄 출력  

```python
import sys
sys.stdin = open("Input/Input_BOJ10989.txt", "r")

N = int(sys.stdin.readline())
ls_cnt = [0] * 10000

## List는 생성하지 않는다.
# ls_num = [int(sys.stdin.readline()) for _ in range(N)]
# ls_num.sort()

for _ in range(N): ls_cnt[int(sys.stdin.readline()) - 1] += 1
    
for i in range(len(ls_cnt)): 
    for _ in range(ls_cnt[i]): sys.stdout.write((str(i + 1) + '\n'))
```
