```python
## N은 10일 때, i별 개수
# 1 * 10
# 2 * 5
# 3 * 3
# 4 * 2
# 5 * 2
# 6 * 1
# 7 * 1
# 8 * 1
# 9 * 1
# 10 * 1

## 해당 Code는 자연수 2로부터 순서쌍을 저장하지만,
## 시간 초과로 인해 오답처리된다.
## 따라서, 약수의 합을 구하는 점화식을 적용해야한다.
while(True):
  try: num, sum = int(input()), 0 ## 변수 입력 및 초기화
  except: break

  for i in range(1, num+1) : sum += (num // i) * i
  print(sum)
```
