```python
## 제곱과 제곱근

num = 2
## 1) 제곱
print(pow(num, 2))

## 2-1) 제곱근
import math
print(math.sqrt(num))

## 2-2) 제곱근(Math Library 사용 불가시)
print(2 ** 0.5)

## 2-3) 루트
a, b, c = 15, 18, 26
## 1 3 5 15
## 1 2 3 6 9 18
## 1 2 13 26
print(15 ** 0.5)
print(18 ** 0.5)
print(26 ** 0.5)

## 3) 승 수 구하기
## 27은 3의 몇 승인가?
num, div, cnt = 27, 3, 0
while num//div != 1:
    cnt+=1
    num//=div
print(cnt)
```

```python
## 입력 받기
## 공백으로 구분된 수열을 리스트형으로 입력 받기
ls_input = list(map(int, input().split()))
print(ls_input)

## TC와 입력값을 1개의 행에서 동시에 입력받는 경우, '좌 -> 우'
T, ls_input = int(input()), list(map(int, input().split()))
for tc in range(T): print(f"{tc}: {ls_input[tc]}")
```
