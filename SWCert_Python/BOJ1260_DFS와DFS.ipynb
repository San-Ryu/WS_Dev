```python
## List를 Value로 갖는 Dict 선언
dict_map = {0: [1]}
## Dict의 Value가 List이므로 Append 가능하다/
dict_map[0].append(2)
print(dict_map)

## in 조건은 Dict / List 모두에 적용 가능하다.
print(0 in dict_map) ## True
print(1 in dict_map) ## False
print(0 in dict_map[0]) ## False
print(1 in dict_map[0]) ## False
```

```python
# 4 5 1
# 1 2
# 1 3
# 1 4
# 2 4
# 3 4

import sys
sys.stdin = open('Input/Input_BOJ1260.txt', 'r')

T = int(sys.stdin.readline())

for tc in range(T):
    print(tc)

n, m, v = sys.stdin.readline().split(' ')
print(n, m ,v)

dict_line = {}
for line in range(int(m)):
    print(line)
    n1, n2 = sys.stdin.readline().split(' ')
    n1, n2 = int(n1), int(n2)

    if (n1 in dict_line) and (n2 in dict_line):
        dict_line[n1].append(n2)
        dict_line[n2].append(n1)
    elif (n1 not in dict_line) and (n2 in dict_line):
        dict_line[n1] = [n2]
        dict_line[n2].append(n1)
    elif (n1 in dict_line) and (n2 not in dict_line):
        dict_line[n1].append(n2)
        dict_line[n2] = [n1]
    elif (n1 not in dict_line) and (n2 not in dict_line):
        dict_line[n1], dict_line[n2] = [n2], [n1]

print(dict_line)
```

```python
# 4 5 1
# 1 2
# 1 3
# 1 4
# 2 4
# 3 4

import sys
from collections import defaultdict, deque
sys.stdin = open('Input/Input_BOJ1260.txt', 'r')

# T = int(sys.stdin.readline())
n, m, v = map(int, sys.stdin.readline().split())

dict_line = defaultdict(list)
for _ in range(m):
    n1, n2 = map(int, sys.stdin.readline().split(' '))
    dict_line[n1].append(n2)
    dict_line[n2].append(n1)

## 노드에 대한 정렬
for key in dict_line: dict_line[key].sort()

## DFS
def dfs(start, visited):
    visited.add(start)
    print(start, end=" ")
    for nxt in dict_line[start]:
        if nxt not in visited: dfs(nxt, visited)

## BFS
def bfs(start):
    visited = set([start])
    q = deque([start])
    while q:
        current = q.popleft()
        print(current, end=" ")
        for nxt in dict_line[current]:
            if nxt not in visited:
                visited.add(nxt)
                q.append(nxt)

## 실행
dfs(v, set())
print()
bfs(v)
```
