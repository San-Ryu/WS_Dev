### BOJ 17425 약수의 합
[링크] https://www.acmicpc.net/problem/17425

[문제]
두 자연수 A와 B가 있을 때, A = BC를 만족하는 자연수 C를 A의 약수라고 한다. 예를 들어, 2의 약수는 1, 2가 있고, 24의 약수는 1, 2, 3, 4, 6, 8, 12, 24가 있다. 자연수 A의 약수의 합은 A의 모든 약수를 더한 값이고, f(A)로 표현한다. x보다 작거나 같은 모든 자연수 y의 f(y)값을 더한 값은 g(x)로 표현한다.

자연수 N이 주어졌을 때, g(N)을 구해보자.

[입력]
첫째 줄에 테스트 케이스의 개수 T(1 ≤ T ≤ 100,000)가 주어진다. 둘째 줄부터 테스트 케이스가 한 줄에 하나씩 주어지며 자연수 N(1 ≤ N ≤ 1,000,000)이 주어진다.

[출력]
각각의 테스트 케이스마다, 한 줄에 하나씩 g(N)를 출력한다.

[예제 입력 1]
5
1
2
10
70
10000

[예제 출력 1]
1
4
87
4065
82256014

[MEMO]
여기서도 마찬가지로,
1) 직접 나누어 구하는 방식은 Timeout으로 판정된다.
2) 약수의 합을 구하는 점화식을 적용하는 방식 또한 Timeout으로 판정된다.
--> DP 배열을 미리 만들어놓고 사용하는 방법을 채택할 수 있다.

그리고, 해당 방법으로도 불가하다면
제출 방식을 python3로부터 pypy3으로 변경하여 제출해봐야한다.

```python
## pypy3 제출시 정답
max_dp = 1000000
dp, s = [1]*(max_dp + 1), [0]*(max_dp + 1)

for i in range(2, max_dp+1):
  j = 1
  for j in range(i, max_dp+1, i):
    dp[j] += i

for i in range(1, max_dp+1):
  s[i] = s[i-1] + dp[i]

n, ans = int(input()), []
for _ in range(n):
  a = int(input())
  ans.append(s[a])
print('\n'.join(map(str, ans))+'\n')
```
