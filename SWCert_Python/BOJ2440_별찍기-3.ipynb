### BOJ2440 별찍기-3
[링크] https://www.acmicpc.net/problem/2440

[문제]
첫째 줄에는 별 N개, 둘째 줄에는 별 N-1개, ..., N번째 줄에는 별 1개를 찍는 문제

[입력]
첫째 줄에 N(1 ≤ N ≤ 100)이 주어진다.

[출력]
첫째 줄부터 N번째 줄까지 차례대로 별을 출력한다.

[예제 입력 1]
5

[예제 출력 1]
###### *****
###### ****
###### ***
###### **
###### *

[MEMO]

```python
num = int(input())

for i in range(num+1, 1, -1): print("*" * (i-1))
## 여기도 마찬가지다.
## range(num+1, 0, -1)으로 지정할 경우, 정답 끝에 빈 줄이 추가된다.
```
