### BOJ1002 터렛  
[링크] https://www.acmicpc.net/problem/1002  

[문제]  
조규현과 백승환은 터렛에 근무하는 직원이다. 하지만 워낙 존재감이 없어서 인구수는 차지하지 않는다.   
  
이석원은 조규현과 백승환에게 상대편 마린(류재명)의 위치를 계산하라는 명령을 내렸다. 조규현과 백승환은 각각 자신의 터렛 위치에서 현재 적까지의 거리를 계산했다.  

조규현의 좌표 $(x_1, y_1)$와 백승환의 좌표 $(x_2, y_2)$가 주어지고, 조규현이 계산한 류재명과의 거리 r_1과 백승환이 계산한 류재명과의 거리 r_2가 주어졌을 때, 류재명이 있을 수 있는 좌표의 수를 출력하는 프로그램을 작성하시오.

[입력]  
첫째 줄에 테스트 케이스의 개수 T가 주어진다. 각 테스트 케이스는 다음과 같이 이루어져 있다.  

한 줄에 공백으로 구분 된 여섯 정수  
$x_1$, $y_1$, $r_1$
, $x_2$, $y_2$, $r_2$  
가 주어진다.

[출력]  
각 테스트 케이스마다 류재명이 있을 수 있는 위치의 수를 출력한다. 만약 류재명이 있을 수 있는 위치의 개수가 무한대일 경우에는 -1 출력한다.  

[제한]  
-10,000 ≤ x_1, y_1, x_2, y_2 ≤ 10,000  
1 ≤ r_1, r_2 ≤ 10,000

[예제 입력 1]  
3  
0 0 13 40 0 37  
0 0 3 0 7 4  
1 1 1 1 1 5  

[예제 출력 1]  
2  
1  
0  

[MEMO]  
두 터렛의 좌표는 각각의 영역에 대한 원의 중심을,
마린과의 거리는 원의 반지름을 의미한다.  
따라서, 두 개의 원의 접점의 개수를 구하는 문제이다.  

문제에서 요구하는 답은 "위치의 좌표"가 아닌, "위치의 수"이다.  
따라서, 아래와 같은 4가지 경우로 나뉜다.
1) 두 원이 접하지 않는 경우  
2) 두 원의 접점이 1개인 경우  
3) 두 원의 접점이 2개인 경우  
4) 두 원의 접점이 무수히 많은 경우 (중심과 반지름이 모두 같음)  

이 점을 착안하여 문제 풀이를 진행해보자

<오답>  
1. 내접/포함 관계 판정 오류  
- 코드에서는 dist < min(r1, r2)일 때 "접하지 않음"이라고 처리했어요.
- 하지만 실제로는 큰 원 안에 작은 원이 들어가 있는 경우를 판정할 때는
dist + min(r1, r2) < max(r1, r2) 조건을 써야 합니다.
- 즉, 중심 간 거리 + 작은 반지름이 큰 반지름보다 작을 때, 작은 원이 큰 원 안에 완전히 들어가 접점이 없는 경우입니다.
- 현재 코드처럼 단순히 dist < min(r1, r2)로 비교하면 잘못된 결과가 나옵니다.

2. 내접 조건 판정 오류  
- 내접은 dist + min(r1, r2) == max(r1, r2)일 때 발생합니다.
- 코드에서는 dist == min(r1, r2)로 판정했는데, 이는 잘못된 조건입니다.
예를 들어, 중심 간 거리가 작은 반지름과 같다고 해서 내접이 되는 것이 아닙니다.



```python
## 입력값: Testcase 개수
T = int(input())

for _ in range(T):
  ## 입력값: 각 중심점의 좌표, 반지름
  x1, y1, r1, x2, y2, r2 = list(map(int, input().split()))
  ## 두 중심간 거리
  dist = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5

  ## 동심원인 경우
  ## 두 반지름이 같아, 접점이 무수히 많은 경우 // 두 반지름이 달라, 접점이 없는 경우
  if (x1 == x2) and (y1 == y2): print(-1) if r1 == r2 else print(0)

  else: ## 두 원의 중심의 위치가 다른 경우
    if dist > r1 + r2: print(0) ## 두 원이 서로 떨어져있는 경우
    elif dist == r1 + r2: print(1) ## 두 원의 접점이 한 개인 경우 01 (외접)
    else: ## dist < r1 + r2인 경우
      if dist + min(r1, r2) < max(r1, r2): print(0) ## 한 원이 다른 원에 들어가있는 경우 (접하지 않음)
      elif dist + min(r1, r2) == max(r1, r2): print(1) ## 두 원이 접점이 한 개인 경우 (내접)
      else: print(2)  ## 두 원의 접점이 두 개인 경우
```

```python
# ## tc 1-1: 동심원, 같은 반지름
# 0 0 5 0 0 5
# -1

# ## tc 1-2: 동심원, 다른 반지름
# 0 0 5 0 0 3
# 0

# ## tc 2-1: 다른 중심, dist == r1 + r2 (외접)
# 0 0 1 2 0 1
# 1

# ## tc 2-2: 다른 중심, dist == abs(r1 - r2) (내접)
# 0 0 2 0 1 1
# 1

# 0 1 1 0 0 2

# ## tc 2-3: 다른 중심, dist > r1 + r2 (서로 바깥에 위치, 접하지 않음)
# 0 0 1 2 2 1
# 0

# ## tc 2-4: 다른 중심, dist < r1 + r2 (한 원이 안쪽에 위치, 접하지 않음)
# 0 0 5 1 0 1
# 0

# 1 0 1 0 0 5
# 0
```

### BOJ10811 바구니 뒤집기  
[링크] https://www.acmicpc.net/problem/10811

도현이는 앞으로 M번 바구니의 순서를 역순으로 만들려고 한다. 도현이는 한 번 순서를 역순으로 바꿀 때, 순서를 역순으로 만들 범위를 정하고, 그 범위에 들어있는 바구니의 순서를 역순으로 만든다.  
바구니의 순서를 어떻게 바꿀지 주어졌을 때, M번 바구니의 순서를 역순으로 만든 다음, 바구니에 적혀있는 번호를 가장 왼쪽 바구니부터 출력하는 프로그램을 작성하시오.  

[입력]  
첫째 줄에 N (1 ≤ N ≤ 100)과 M (1 ≤ M ≤ 100)이 주어진다.  
둘째 줄부터 M개의 줄에는 바구니의 순서를 역순으로 만드는 방법이 주어진다. 방법은 i j로 나타내고, 왼쪽으로부터 i번째 바구니부터 j번째 바구니의 순서를 역순으로 만든다는 뜻이다. (1 ≤ i ≤ j ≤ N)  
도현이는 입력으로 주어진 순서대로 바구니의 순서를 바꾼다.  

[출력]  
모든 순서를 바꾼 다음에, 가장 왼쪽에 있는 바구니부터 바구니에 적혀있는 순서를 공백으로 구분해 출력한다.  

[예제 입력 1]  
5 4  
1 2  
3 4  
1 4  
2 2  

[예제 출력 1]  
3 4 1 2 5  

[MEMO]  
deque를 이용하면 편리하게 풀 수 있다.

```python
from collections import deque

cnt_num, cnt_sort = map(int, input().split())

ls_num = [num for num in range(1, cnt_num + 1)]

for i in range(0, cnt_sort):
  a, b = map(int, input().split())
  q = deque([num_ord for num_ord in range(a, b + 1)])
  # print(len(q), q)

  for j in range(0, len(q) // 2):
    num_head, num_tail = q.popleft(), q.pop()
    num_temp = ls_num[num_head - 1]
    ls_num[num_head - 1] = ls_num[num_tail - 1]
    ls_num[num_tail - 1] = num_temp

for num in ls_num: print(num, end=' ')
```

[MEMO]  
기존의 방식으로는 문제가 발생한다.  
(1) 메모리 절감을 위한 저장 없는 풀이 : 시간 초과  
(1) 호출 시간 향상을 위한 리스트 저장 풀이 : 메모리 초과  
--> "1000000으로 나눈 나머지"라는 것에 착안하여,  
피사노 주기, 빠른 피보나치를 구현해야 한다.

```python
# ## 저장 없는 풀이 : 14초 정도 소요되어, 시간 초과
# num_fibo00, num_fibo01, num = 0, 1, int(input())

# if num < 2: print(num%1000000)
# else:
#   num_prev01, num_prev02, num_ans = 0, 1, 0 ## 초기의 Prev01에는 fibo00, Prev02에는 fibo01 저장
#   for i in range(0, num - 1):
#     num_ans = num_prev01 + num_prev02
#     num_prev01, num_prev02 = num_prev02, num_ans
#   print(num_ans%1000000)

# ## 리스트 저장 형식 풀이 : Colab은 메모리 초과로 인한 세션 재실행 발생, 메모리 초과
# ls_fibo, num = [0, 1], int(input())

# if num < 2: print(ls_fibo[num]%1000000)
# else:
#   for i in range(0, num - 1): ls_fibo.append(ls_fibo[i] + ls_fibo[i + 1])
#   print(ls_fibo[num]%1000000)
# ## 리스트를 다른 기능 등에 재활용할 것이 아니라면, 비효율적
```

```python
## 메모이제이션 적용 전
def fibo(num):
  if num == 0: return (0, 1) ## 0번째, 1번째
  else:
    f_k, f_kp1 = fibo(num // 2)

    f_2k = f_k * (2 * f_kp1 - f_k)
    f_2kp1 = f_k ** 2 + f_kp1 ** 2

    if num % 2 == 0: return (f_2k, f_2kp1)
    else: return (f_2kp1, f_2k + f_2kp1)

ls_fibo, num_div, num = [0, 1], 1000000, int(input())

print(fibo(num)[0] % num_div)
```

```python
## math library 사용
import math

num = int(input())
k = int(math.log(num, 3)) ## 승수
# print(num, k)

pt_a, pt_b = "***", "* *"
memo_pattern = {1 : [pt_a, pt_b]}

def def_pattern(num):
  k = int(math.log(num, 3))
  if k in memo_pattern:
    return memo_pattern[k]
  else:
    pt_a, pt_b = def_pattern(num // 3)
    memo_pattern[k] = [pt_a * 3, pt_b * 3]
    return memo_pattern[k]


pt_a, pt_b = def_pattern(num)
print(memo_pattern)
print(f"{pt_a}\n{pt_b}\n{pt_a}")
```

```python
## math library 미사용
def draw_star(num):
  if num == 1: return ["*"]

  sub_pattern = draw_star(num//3)
  print(num, "sub_pattern")
  print(sub_pattern)
  result = []

  for line in sub_pattern: result.append(line * 3)
  for line in sub_pattern: result.append(line + " " * (num // 3) + line)
  for line in sub_pattern: result.append(line * 3)

  return result

## 실행
num = int(input())
pattern = draw_star(num)
print("Pattern", len(pattern))
print(pattern)
print("결과값")
print("\n".join(pattern))
```
