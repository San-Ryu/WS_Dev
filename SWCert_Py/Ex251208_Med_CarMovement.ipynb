[문제]  
N * N의 지도가 있다.
E개의 도로가 있으며, V개의 차가 있다.

(1) 도로
- 입력값에 E개의 도로로서 입력된다.
- 도로는 서로 인접하거나, 지도의 변에 접하지 않는다.
- 도로는 우측 주행 도로이다.

(2) 차량
- 입력값에 V개의 차량으로서 입력된다.
- 각 차는 출발지로부터 최소 움직임으로 주행한다.
- 각 차량별 목표로 하는 좌표가 겹칠 경우, 순번이 가장 낮은 차량이 우선적으로 통과한다.
- 모든 차량이 목적지에 도착하는 경우가 보장된다.
