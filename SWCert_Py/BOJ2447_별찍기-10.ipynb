[문제]  
재귀적인 패턴으로 별을 찍어 보자. N이 3의 거듭제곱(3, 9, 27, ...)이라고 할 때, 크기 N의 패턴은 N×N 정사각형 모양이다.  

크기 3의 패턴은 가운데에 공백이 있고, 가운데를 제외한 모든 칸에 별이 하나씩 있는 패턴이다.  

###### ***  
###### * *  
###### ***  
N이 3보다 클 경우, 크기 N의 패턴은 공백으로 채워진 가운데의 (N/3)×(N/3) 정사각형을 크기 N/3의 패턴으로 둘러싼 형태이다. 예를 들어 크기 27의 패턴은 예제 출력 1과 같다.  

[입력]  
첫째 줄에 N이 주어진다. N은 3의 거듭제곱이다. 즉 어떤 정수 k에 대해 N=3k이며, 이때 1 ≤ k < 8이다.  

[출력]  
첫째 줄부터 N번째 줄까지 별을 출력한다.  

[예제 입력 1]  
27

```python
## math library 사용
import math

num = int(input())
k = int(math.log(num, 3)) ## 승수
# print(num, k)

pt_a, pt_b = "***", "* *"
memo_pattern = {1 : [pt_a, pt_b]}

def def_pattern(num):
  k = int(math.log(num, 3))
  if k in memo_pattern:
    return memo_pattern[k]
  else:
    pt_a, pt_b = def_pattern(num // 3)
    memo_pattern[k] = [pt_a * 3, pt_b * 3]
    return memo_pattern[k]


pt_a, pt_b = def_pattern(num)
print(memo_pattern)
print(f"{pt_a}\n{pt_b}\n{pt_a}")
```

```python
## math library 미사용
def draw_star(num):
  if num == 1: return ["*"]

  sub_pattern = draw_star(num//3)
  print(num, "sub_pattern")
  print(sub_pattern)
  result = []

  for line in sub_pattern: result.append(line * 3)
  for line in sub_pattern: result.append(line + " " * (num // 3) + line)
  for line in sub_pattern: result.append(line * 3)

  return result

## 실행
num = int(input())
pattern = draw_star(num)
print("Pattern", len(pattern))
print(pattern)
print("결과값")
print("\n".join(pattern))
```
