```python
## 해당 문제에서는 별도의 Test case 수 없이, 입력값의 수가 가변적으로 주어진다.
## 따라서, 전체 테스트 케이스의 수를 입력하는게 아니라,
## 더이상 입력값이 들어오지 않으면 반복문을 중지해야한다.
## (런타임 에러 발생)

## 참고로 입력이 되지 않는 경우 ValueError가 발생하나,
## 입력 단에서 발생하는 모든 Exception을 처리할 수 있도록 모든 Exception에 대해 예외처리를 할 것이다.
while(True):
  try : int_n = int(input())
  except : break

  ## 입력값이 2 또는 5로 나누어 떨어지는 경우, 무한루프에 빠지게 된다.
  if (int_n%2 == 0) or (int_n%5 == 0) : continue

  num, int_cnt = 1, 1 ## 나눌 수(1로만 이루어진), 해당 자릿수
  while(True):
    if num%int_n != 0 : num, int_cnt = num*10 + 1, int_cnt + 1
    elif num%int_n == 0 :
      print(int_cnt)
      break
```
