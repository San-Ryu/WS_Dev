### BOJ6588 골드바흐의 추측
[링크] https://www.acmicpc.net/problem/6588

[문제]
1742년, 독일의 아마추어 수학가 크리스티안 골드바흐는 레온하르트 오일러에게 다음과 같은 추측을 제안하는 편지를 보냈다.
"4보다 큰 모든 짝수는 두 홀수 소수의 합으로 나타낼 수 있다."
예를 들어 8은 3 + 5로 나타낼 수 있고, 3과 5는 모두 홀수인 소수이다. 또, 20 = 3 + 17 = 7 + 13, 42 = 5 + 37 = 11 + 31 = 13 + 29 = 19 + 23 이다.
이 추측은 아직도 해결되지 않은 문제이다.
백만 이하의 모든 짝수에 대해서, 이 추측을 검증하는 프로그램을 작성하시오.

[입력]
입력은 하나 또는 그 이상의 테스트 케이스로 이루어져 있다. 테스트 케이스의 개수는 100,000개를 넘지 않는다.
각 테스트 케이스는 짝수 정수 n 하나로 이루어져 있다. (6 ≤ n ≤ 1000000)
입력의 마지막 줄에는 0이 하나 주어진다.

[출력]
각 테스트 케이스에 대해서, n = a + b 형태로 출력한다. 이때, a와 b는 홀수 소수이다. 숫자와 연산자는 공백 하나로 구분되어져 있다. 만약, n을 만들 수 있는 방법이 여러 가지라면, b-a가 가장 큰 것을 출력한다. 또, 두 홀수 소수의 합으로 n을 나타낼 수 없는 경우에는 "Goldbach's conjecture is wrong."을 출력한다.

[예제 입력 1]
8
20
42
0

[예제 출력 1]
8 = 3 + 5
20 = 3 + 17
42 = 5 + 37

```python
## 1안 : n의 최대 범위 내 소수값 리스트를 미리 생성하여 처리
list_oddprime = [2, 3] ## 앞에서 짝수와 3의 배수를 제외하는 과정에서 유실되므로
for i in range(2, 1000001):
  if i%2 ==0 or i%3 == 0: continue ## 짝수와 3의 배수 제외
  for j in range(2, int(i**0.5) + 1):
    if i%j == 0:
      break
  else: list_oddprime.append(i)

num = int(input())
for odd in list_oddprime:
  # print(odd, num-odd)
  # print(num-odd in list_oddsosu)
  if odd > num : continue ## 두 odd의 합을 뽑는 것이므로
  if num-odd in list_oddprime:
    print(f'{num} = {odd} + {num-odd}')
    break
else: print("Goldbach's conjecture is wrong.")
```

```python
import sys

## 2안 : n의 최대 범위 내 소수 여부 리스트를 생성하여 처리
list_isprime = [False] * (1000000 + 1)

## 0과 1은 소수가 아니며, 2와 3은 소수임
list_isprime[0] = list_isprime[1] = False
list_isprime[2] = list_isprime[3] = True

for i in range(2, 1000001):
  if i%2==0 or i%3==0: continue ## 짝수와 3의 배수 False
  for j in range(2, int(i**0.5)+1):
    if i%j==0: break## 약수가 존재할 경우
  else: list_isprime[i] = True

while True:
  try:
    num = int(input())
    if (num < 6) or (num%2!=0): ## 6이하이거나, 홀수인 경우
      print("Goldbach's conjecture is wrong.")
      continue
    for odd in range(2, num):
      ## odd와 num-odd 모두 소수이면
      if num>=6 and (list_isprime[odd] and list_isprime[num - odd]):
        print(f'{num} = {odd} + {num-odd}')
        break
    ## 아닌 경우 (For문을 모두 수행)
    else: print("Goldbach's conjecture is wrong.")
  except EOFError:
    break
```
