### BOJ2442 별찍기-5  
[링크] https://www.acmicpc.net/problem/2442  

[문제]  
첫째 줄에는 별 1개, 둘째 줄에는 별 3개, ..., N번째 줄에는 별 2×N-1개를 찍는 문제  
별은 가운데를 기준으로 대칭이어야 한다.   

[입력]  
첫째 줄에 N(1 ≤ N ≤ 100)이 주어진다.  

[출력]  
첫째 줄부터 N번째 줄까지 차례대로 별을 출력한다.  

[예제 입력 1]  
5  

[예제 출력 1]  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;\*  
&nbsp;&nbsp;&nbsp;&nbsp;\***  
&nbsp;&nbsp;&nbsp;\*****  
&nbsp;&nbsp;\*******  
&nbsp;\*********  

[MEMO]  
num = 5  
num-i(1) / j(2n - 1, 1)  
num-i(2) / j(2n - 1, 3)  
num-i(3) / j(2n - 1, 5)  
num-i(4) / j(2n - 1, 7)  
num-i(5) / j(2n - 1, 9)  


4 / 1  
3 / 3  
2 / 5  
1 / 7  
0 / 9  


```python
num = int(input())

for i in range(1, num+1): print(" " * (num - i) + "*" * (2 * i - 1))
```
