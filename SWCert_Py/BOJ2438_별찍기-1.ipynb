### BOJ2438 별찍기-1
[링크] https://www.acmicpc.net/problem/2438

[문제]
첫째 줄에는 별 1개, 둘째 줄에는 별 2개, N번째 줄에는 별 N개를 찍는 문제

[입력]
첫째 줄에 N(1 ≤ N ≤ 100)이 주어진다.

[출력]
첫째 줄부터 N번째 줄까지 차례대로 별을 출력한다.

[예제 입력 1]
5

[예제 출력 1]
###### *
###### **
###### ***
###### ****
###### *****

[MEMO]

```python
## 이 코드는 정답이고
num = int(input())

for i in range(0,  num): print("*" * (i+1))
```

```python
## 이 코드는 오답이다.
num = int(input())

for i in range(num + 1): print("*" * i)
```
