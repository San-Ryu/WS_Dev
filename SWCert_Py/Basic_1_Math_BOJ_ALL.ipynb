### BOJ4375 1
[링크] https://www.acmicpc.net/problem/4375  

[문제]  
2와 5로 나누어 떨어지지 않는 정수 n(1 ≤ n ≤ 10000)가 주어졌을 때, 각 자릿수가 모두 1로만 이루어진 n의 배수를 찾는 프로그램을 작성하시오.

[입력]  
입력은 여러 개의 테스트 케이스로 이루어져 있다. 각 테스트 케이스는 한 줄로 이루어져 있고, n이 주어진다.

[출력]  
각 자릿수가 모두 1로만 이루어진 n의 배수 중 가장 작은 수의 자리수를 출력한다.

[예제 입력 1]  
3  
7  
9901  

[예제 출력 1]  
3  
6  
12  

[MEMO]  
"문제"에서의 n의 배수를 찾는 것이 아니라,  
각 자릿수가 모두 1로 이루어진 n의 배수 중 가장 작은 자리수를 출력하는 것이다.  


```python
# ## 전체 테스트 케이스
# int_T = 3

# for tc in range(0, int_T):
#   int_n = int(input())

#   num, int_cnt = 1, 1
#   while(True):
#     if num%int_n != 0 : num, int_cnt = num*10 + 1, int_cnt + 1
#     elif num%int_n == 0 :
#       print(int_cnt)
#       break
```

```python
## 해당 문제에서는 별도의 Test case 수 없이, 입력값의 수가 가변적으로 주어진다.
## 따라서, 전체 테스트 케이스의 수를 입력하는게 아니라,
## 더이상 입력값이 들어오지 않으면 반복문을 중지해야한다.
## (런타임 에러 발생)

## 참고로 입력이 되지 않는 경우 ValueError가 발생하나,
## 입력 단에서 발생하는 모든 Exception을 처리할 수 있도록 모든 Exception에 대해 예외처리를 할 것이다.
while(True):
  try : int_n = int(input())
  except : break

  ## 입력값이 2 또는 5로 나누어 떨어지는 경우, 무한루프에 빠지게 된다.
  if (int_n%2 == 0) or (int_n%5 == 0) : continue

  num, int_cnt = 1, 1 ## 나눌 수(1로만 이루어진), 해당 자릿수
  while(True):
    if num%int_n != 0 : num, int_cnt = num*10 + 1, int_cnt + 1
    elif num%int_n == 0 :
      print(int_cnt)
      break
```

### BOJ1037 약수  
[링크] https://www.acmicpc.net/problem/1037  

[문제]  
양수 A가 N의 진짜 약수가 되려면, N이 A의 배수이고, A가 1과 N이 아니어야 한다. 어떤 수 N의 진짜 약수가 모두 주어질 때, N을 구하는 프로그램을 작성하시오.

[입력]  
첫째 줄에 N의 진짜 약수의 개수가 주어진다. 이 개수는 50보다 작거나 같은 자연수이다. 둘째 줄에는 N의 진짜 약수가 주어진다. 1,000,000보다 작거나 같고, 2보다 크거나 같은 자연수이고, 중복되지 않는다.

[출력]  
첫째 줄에 N을 출력한다. N은 항상 32비트 부호있는 정수로 표현할 수 있다.

[예제 입력 1]

[예제 출력 4]  
185192

[MEMO]  
"1과 N을 제외한 모든 진짜 약수"가 힌트이다.  
크기에 따라 정렬된 진짜 약수 리스트의 가장 첫번째 값과 마지막 값을 곱하면 N이 나올 수 있다.  
단, 아래와 같은 이유로 Min값과 Max값을 곱하는 방식으로 풀이해야 한다.  
(1) 4, 9와 같이 진짜 약수가 1개인 경우가 존재   
(2) 입력값을 크기에 따라 정렬하는 데에 소모되는 자원 절약  

```python
while(True):
  try: list_inp, len_inp = [], int(input())
  except: break

  list_inp = list(map(int, input().split()))

  int_min, int_max = list_inp[0], list_inp[0]
  for inp in list_inp:
    if int_min > inp: int_min = inp ## 최소값 갱신
    if int_max < inp: int_max = inp ## 최댓값 갱신
  print(int_min * int_max)
```

```python
## pypy3 제출시 정답
max_dp = 1000000
dp, s = [1]*(max_dp + 1), [0]*(max_dp + 1)

for i in range(2, max_dp+1):
  j = 1
  for j in range(i, max_dp+1, i):
    dp[j] += i

for i in range(1, max_dp+1):
  s[i] = s[i-1] + dp[i]

n, ans = int(input()), []
for _ in range(n):
  a = int(input())
  ans.append(s[a])
print('\n'.join(map(str, ans))+'\n')
```

### BOJ2609 최대공약수와 최소공배수
[링크] https://www.acmicpc.net/problem/2609  

[문제]  
두 개의 자연수를 입력받아 최대 공약수와 최소 공배수를 출력하는 프로그램을 작성하시오.

[입력]  
첫째 줄에는 두 개의 자연수가 주어진다. 이 둘은 10,000이하의 자연수이며 사이에 한 칸의 공백이 주어진다.

[출력]  
첫째 줄에는 입력으로 주어진 두 수의 최대공약수를, 둘째 줄에는 입력으로 주어진 두 수의 최소 공배수를 출력한다.

[예제 입력 1]  
24 18  

[예제 출력 1]  
6 72  

[MEMO]  
유클리드 호재법을 통해 GGC(최대공약수)를, 최대공약수를 기반으로 LCM(최소공배수)를 산출한다.

```python
def gcd(a, b):
  while b > 0: a, b = b, a % b
  return a

def lcm(a, b): return int((a * b) / gcd(a, b)) ## int()로 감싸주지 않으면, float로 출력되어 오답

a, b = map(int, input().split())
print(gcd(a, b))
print(lcm(a, b))
```

3  

[MEMO]  


```python
n, ls = int(input()), list(map(int, input().split()))

count = 0
for num in ls:
  check = 1
  for i in range(2, num):
    if num % i == 0:
      check = 0
      break
  ## check = 1(나누어떨어지지 않음) and num!=1 (1이 아닌 수)
  if check and num != 1: count += 1
print(count)
```

```python
## 1안 : n의 최대 범위 내 소수값 리스트를 미리 생성하여 처리
list_oddprime = [2, 3] ## 앞에서 짝수와 3의 배수를 제외하는 과정에서 유실되므로
for i in range(2, 1000001):
  if i%2 ==0 or i%3 == 0: continue ## 짝수와 3의 배수 제외
  for j in range(2, int(i**0.5) + 1):
    if i%j == 0:
      break
  else: list_oddprime.append(i)

num = int(input())
for odd in list_oddprime:
  # print(odd, num-odd)
  # print(num-odd in list_oddsosu)
  if odd > num : continue ## 두 odd의 합을 뽑는 것이므로
  if num-odd in list_oddprime:
    print(f'{num} = {odd} + {num-odd}')
    break
else: print("Goldbach's conjecture is wrong.")
```

```python
import sys

## 2안 : n의 최대 범위 내 소수 여부 리스트를 생성하여 처리
list_isprime = [False] * (1000000 + 1)

## 0과 1은 소수가 아니며, 2와 3은 소수임
list_isprime[0] = list_isprime[1] = False
list_isprime[2] = list_isprime[3] = True

for i in range(2, 1000001):
  if i%2==0 or i%3==0: continue ## 짝수와 3의 배수 False
  for j in range(2, int(i**0.5)+1):
    if i%j==0: break## 약수가 존재할 경우
  else: list_isprime[i] = True

while True:
  try:
    num = int(input())
    if (num < 6) or (num%2!=0): ## 6이하이거나, 홀수인 경우
      print("Goldbach's conjecture is wrong.")
      continue
    for odd in range(2, num):
      ## odd와 num-odd 모두 소수이면
      if num>=6 and (list_isprime[odd] and list_isprime[num - odd]):
        print(f'{num} = {odd} + {num-odd}')
        break
    ## 아닌 경우 (For문을 모두 수행)
    else: print("Goldbach's conjecture is wrong.")
  except EOFError:
    break
```
