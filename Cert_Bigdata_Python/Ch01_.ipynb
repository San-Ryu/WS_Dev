```python
import kagglehub

# Download latest version
## LOL 전적 데이터
# path = kagglehub.dataset_download("datasnaek/league-of-legends")
## 우울증 환자 데이터
path = kagglehub.dataset_download("anthonytherrien/depression-dataset")

print("Path to dataset files:", path)
```

```python
## LOL 전적 데이터
# df = pd.read_csv(path + '\games.csv', encoding='cp949')
## 우울증 환자 데이터
df = pd.read_csv(path + '\depression_data.csv', encoding='cp949')
```

```python
## 평균
print(df[df["Education Level"]=="Master's Degree"]["Age"].mean())
## 중앙값
print(df[df["Education Level"]=="Master's Degree"]["Age"].median())
## 최빈값
print(df[df["Education Level"]=="Master's Degree"]["Age"].mode())
```
